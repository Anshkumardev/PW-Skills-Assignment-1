Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px

df = px.data.tips()

fig = px.histogram(df, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                   title="Histogram of Total Bills by Sex, Colored by Day and Patterned by Smoker Status")

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot, or distribution plot, visualizes the distribution of a dataset across a single variable by combining a histogram with a kernel density estimate (KDE) or rug plot. In Plotly Express, similar effects can be achieved through histograms and separate KDE calculations, as it lacks a built-in distplot function.

```python
import plotly.express as px

df = px.data.tips()

fig = px.histogram(df, x="total_bill",
                   marginal="rug", 
                   title="Distribution of Total Bills")
fig.show()
```
