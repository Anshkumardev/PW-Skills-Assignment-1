Q4: Explain Bayes's theorem with an example.

# Bayes's Theorem

Bayes's theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event. It is formulated as:

\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

Where:
- \( P(A|B) \) is the probability of event A occurring given that B is true.
- \( P(B|A) \) is the probability of event B occurring given that A is true.
- \( P(A) \) is the probability of event A.
- \( P(B) \) is the probability of event B.

## Example:
Suppose you want to determine the probability that a person has a disease (D) given that they tested positive (T) for it. You know:

- \( P(D) = 0.01 \) (1% of the population has the disease)
- \( P(T|D) = 0.99 \) (the test is 99% accurate for those with the disease)
- \( P(T|\neg D) = 0.05 \) (the test has a 5% false positive rate for those without the disease)
- \( P(\neg D) = 0.99 \) (99% of the population does not have the disease)

First, calculate \( P(T) \):
\[ P(T) = P(T|D) \cdot P(D) + P(T|\neg D) \cdot P(\neg D) \]
\[ P(T) = 0.99 \cdot 0.01 + 0.05 \cdot 0.99 \]
\[ P(T) = 0.0099 + 0.0495 \]
\[ P(T) = 0.0594 \]

Now, apply Bayes's theorem:
\[ P(D|T) = \frac{P(T|D) \cdot P(D)}{P(T)} \]
\[ P(D|T) = \frac{0.99 \cdot 0.01}{0.0594} \]
\[ P(D|T) \approx 0.166 \]

So, the probability that a person has the disease given that they tested positive is approximately 16.6%.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

## Confidence Interval

A confidence interval (CI) is a range of values used to estimate the true value of a population parameter. It provides an interval within which we are confident that the parameter lies, based on sample data. The confidence level (e.g., 95%) indicates the probability that the interval contains the true parameter.

### Calculation of Confidence Interval

1. **Calculate the sample mean (\( \bar{x} \)):**
   \[
   \bar{x} = \frac{\sum{x_i}}{n}
   \]
   where \( x_i \) are the sample observations and \( n \) is the sample size.

2. **Find the standard deviation (s) of the sample:**
   \[
   s = \sqrt{\frac{\sum{(x_i - \bar{x})^2}}{n-1}}
   \]

3. **Determine the standard error (SE):**
   \[
   SE = \frac{s}{\sqrt{n}}
   \]

4. **Choose the confidence level and find the corresponding z-score (for 95%, z ≈ 1.96):**

5. **Calculate the margin of error (ME):**
   \[
   ME = z \times SE
   \]

6. **Compute the confidence interval:**
   \[
   \text{CI} = (\bar{x} - ME, \bar{x} + ME)
   \]

### Example

Suppose we have a sample of 30 students' test scores with a mean score of 80 and a standard deviation of 10. We want to calculate the 95% confidence interval for the population mean.

1. **Sample mean (\( \bar{x} \)):** 80
2. **Standard deviation (s):** 10
3. **Sample size (n):** 30
4. **Standard error (SE):**
   \[
   SE = \frac{10}{\sqrt{30}} \approx 1.83
   \]
5. **Z-score for 95% confidence level:** 1.96
6. **Margin of error (ME):**
   \[
   ME = 1.96 \times 1.83 \approx 3.59
   \]
7. **Confidence interval (CI):**
   \[
   CI = (80 - 3.59, 80 + 3.59) = (76.41, 83.59)
   \]

Thus, we are 95% confident that the true mean score of the population lies between 76.41 and 83.59.

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

## Bayes' Theorem

Bayes' Theorem is used to calculate the probability of an event occurring based on prior knowledge of conditions related to the event. It is expressed as:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:
- \( P(A|B) \) is the probability of event A occurring given event B has occurred.
- \( P(B|A) \) is the probability of event B occurring given event A has occurred.
- \( P(A) \) is the prior probability of event A.
- \( P(B) \) is the prior probability of event B.

### Sample Problem

A medical test for a disease is 99% accurate. This means:
- The probability of a positive test given the person has the disease \( P(Positive | Disease) = 0.99 \)
- The probability of a negative test given the person does not have the disease \( P(Negative | No Disease) = 0.99 \)

The disease is rare, affecting 1% of the population:
- \( P(Disease) = 0.01 \)
- \( P(No Disease) = 0.99 \)

We want to find the probability that a person has the disease given a positive test result, \( P(\text{Disease} | \text{Positive}) \).

### Solution

1. **Calculate \( P(\text{Positive}) \):**
   \[
   P(\text{Positive}) = P(\text{Positive} | \text{Disease}) \cdot P(\text{Disease}) + P(\text{Positive} | \text{No Disease}) \cdot P(\text{No Disease})
   \]
   \[
   P(\text{Positive}) = 0.99 \cdot 0.01 + 0.01 \cdot 0.99 = 0.0099 + 0.0099 = 0.0198
   \]

2. **Apply Bayes' Theorem:**
   \[
   P(\text{Disease} | \text{Positive}) = \frac{P(\text{Positive} | \text{Disease}) \cdot P(\text{Disease})}{P(\text{Positive})}
   \]
   \[
   P(\text{Disease} | \text{Positive}) = \frac{0.99 \cdot 0.01}{0.0198} \approx 0.5
   \]

Thus, the probability that a person has the disease given a positive test result is approximately 50%.

Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation

## 95% Confidence Interval Calculation

To calculate the 95% confidence interval for a sample with a mean of 50 and a standard deviation of 5, follow these steps:

1. **Sample Mean (\( \bar{x} \))**: 50
2. **Standard Deviation (s)**: 5
3. **Sample Size (n)**: Assume n = 30 (if not given, it's typical to assume a reasonably large sample size for normal distribution).

4. **Standard Error (SE)**:
   \[
   SE = \frac{s}{\sqrt{n}} = \frac{5}{\sqrt{30}} \approx 0.91
   \]

5. **Z-score for 95% confidence level**: 1.96

6. **Margin of Error (ME)**:
   \[
   ME = z \times SE = 1.96 \times 0.91 \approx 1.78
   \]

7. **Confidence Interval (CI)**:
   \[
   CI = (\bar{x} - ME, \bar{x} + ME) = (50 - 1.78, 50 + 1.78) = (48.22, 51.78)
   \]

### Interpretation

We are 95% confident that the true population mean lies between 48.22 and 51.78. This means that if we were to take many samples and calculate a confidence interval for each, approximately 95% of these intervals would contain the true population mean.

Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

## Margin of Error in a Confidence Interval

The margin of error (ME) is the amount added and subtracted from the sample mean to create the confidence interval. It reflects the uncertainty in the estimate of the population parameter.

\[
ME = z \times \frac{s}{\sqrt{n}}
\]

where:
- \( z \) is the z-score for the desired confidence level.
- \( s \) is the sample standard deviation.
- \( n \) is the sample size.

### Effect of Sample Size on Margin of Error

As the sample size (\( n \)) increases, the margin of error decreases. This is because the standard error (\( \frac{s}{\sqrt{n}} \)) decreases with larger \( n \), making the estimate more precise.

### Example Scenario

Estimating the average height of adult men in a city:

- **Small Sample Size (n=30):**
  - Sample mean (\( \bar{x} \)): 175 cm
  - Standard deviation (\( s \)): 10 cm
  - Confidence level: 95% (z-score ≈ 1.96)
  - Margin of error:
    \[
    ME = 1.96 \times \frac{10}{\sqrt{30}} \approx 3.58
    \]
  - Confidence interval: \( 175 \pm 3.58 \) (171.42 cm to 178.58 cm)

- **Larger Sample Size (n=100):**
  - Sample mean (\( \bar{x} \)): 175 cm
  - Standard deviation (\( s \)): 10 cm
  - Confidence level: 95% (z-score ≈ 1.96)
  - Margin of error:
    \[
    ME = 1.96 \times \frac{10}{\sqrt{100}} = 1.96
    \]
  - Confidence interval: \( 175 \pm 1.96 \) (173.04 cm to 176.96 cm)

Increasing the sample size from 30 to 100 reduces the margin of error from 3.58 cm to 1.96 cm, resulting in a more precise estimate.

Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.

### Calculating the Z-Score

The z-score is calculated using the formula:

\[
z = \frac{(X - µ)}{\sigma}
\]

where:
- \( X \) is the data point (75)
- \(µ) is the population mean (70)
- \( \sigma \) is the population standard deviation (5)

Substitute the values:

\[
z = \frac{(75 - 70)}{5} = \frac{5}{5} = 1
\]

### Interpretation

A z-score of 1 means the data point (75) is 1 standard deviation above the population mean (70).

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

## Hypothesis Test for Weight Loss Drug Effectiveness

We will conduct a t-test to determine if the drug is significantly effective at a 95% confidence level.

### Given Data

- Sample size (\( n \)): 50
- Sample mean (\( \bar{x} \)): 6 pounds
- Sample standard deviation (\( s \)): 2.5 pounds
- Population mean (\( \mu \)): 0 (null hypothesis)

### Hypotheses

- Null hypothesis (\( H_0 \)): The drug is not effective (\( \mu = 0 \)).
- Alternative hypothesis (\( H_1 \)): The drug is effective (\( \mu > 0 \)).

### T-Test Calculation

1. **Calculate the t-score:**
   \[
   t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
   \]
   \[
   t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} = \frac{6}{0.3536} \approx 16.97
   \]

2. **Degrees of freedom (df):**
   \[
   df = n - 1 = 50 - 1 = 49
   \]

3. **Determine the critical t-value for a one-tailed test at 95% confidence level:**
   \[
   t_{critical} \approx 1.676 (from t-distribution table for df = 49)
   \]

4. **Compare t-score with critical t-value:**
   \[
   t = 16.97 > t_{critical} = 1.676
   \]

### Conclusion

Since the calculated t-score (16.97) is greater than the critical t-value (1.676), we reject the null hypothesis. This indicates that the weight loss drug is significantly effective at the 95% confidence level.

Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.

## Confidence Interval for Job Satisfaction

In a survey of 500 people, 65% reported being satisfied with their current job. We will calculate the 95% confidence interval for the true proportion.

### Given Data

- Sample size (\( n \)): 500
- Sample proportion (\( \hat{p} \)): 0.65
- Confidence level: 95% (z-score ≈ 1.96)

### Confidence Interval Calculation

1. **Calculate the standard error (SE):**
   \[
   SE = \sqrt{\frac{\hat{p} (1 - \hat{p})}{n}}
   \]
   \[
   SE = \sqrt{\frac{0.65 \times 0.35}{500}} \approx 0.0214
   \]

2. **Calculate the margin of error (ME):**
   \[
   ME = z \times SE
   \]
   \[
   ME = 1.96 \times 0.0214 \approx 0.0419
   \]

3. **Compute the confidence interval:**
   \[
   CI = \hat{p} \pm ME
   \]
   \[
   CI = 0.65 \pm 0.0419
   \]
   \[
   CI = (0.6081, 0.6919)
   \]

### Conclusion

The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately (0.6081, 0.6919).

Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.

## Hypothesis Test for Comparing Two Teaching Methods

We will conduct a t-test to determine if there is a significant difference in student performance between two teaching methods at a 0.01 significance level.

### Given Data

- **Sample A:**
  - Mean (\( \bar{x}_A \)): 85
  - Standard deviation (\( s_A \)): 6
  - Sample size (\( n_A \)): 30 (assumed)

- **Sample B:**
  - Mean (\( \bar{x}_B \)): 82
  - Standard deviation (\( s_B \)): 5
  - Sample size (\( n_B \)): 30 (assumed)

### Hypotheses

- Null hypothesis (\( H_0 \)): There is no significant difference between the teaching methods (\( \mu_A = \mu_B \)).
- Alternative hypothesis (\( H_1 \)): There is a significant difference between the teaching methods (\( \mu_A \neq \mu_B \)).

### T-Test Calculation

1. **Calculate the pooled standard deviation ( \( s_p \) ):**
   \[
   s_p = \sqrt{\frac{(n_A - 1)s_A^2 + (n_B - 1)s_B^2}{n_A + n_B - 2}}
   \]
   \[
   s_p = \sqrt{\frac{(30 - 1)6^2 + (30 - 1)5^2}{30 + 30 - 2}} = \sqrt{\frac{29 \cdot 36 + 29 \cdot 25}{58}} \approx \sqrt{\frac{1044 + 725}{58}} \approx \sqrt{30.56} \approx 5.53
   \]

2. **Calculate the standard error (SE) of the difference:**
   \[
   SE = s_p \sqrt{\frac{1}{n_A} + \frac{1}{n_B}}
   \]
   \[
   SE = 5.53 \sqrt{\frac{1}{30} + \frac{1}{30}} = 5.53 \sqrt{\frac{2}{30}} \approx 5.53 \cdot 0.2582 \approx 1.43
   \]

3. **Calculate the t-score:**
   \[
   t = \frac{\bar{x}_A - \bar{x}_B}{SE}
   \]
   \[
   t = \frac{85 - 82}{1.43} \approx 2.10
   \]

4. **Degrees of freedom (df):**
   \[
   df = n_A + n_B - 2 = 30 + 30 - 2 = 58
   \]

5. **Determine the critical t-value for a two-tailed test at a 0.01 significance level:**
   \[
   t_{critical} \approx 2.660 (from t-distribution table for df = 58)
   \]

6. **Compare t-score with critical t-value:**
   \[
   t = 2.10 < t_{critical} = 2.660
   \]

### Conclusion

Since the calculated t-score (2.10) is less than the critical t-value (2.660), we fail to reject the null hypothesis. This indicates that there is no significant difference in student performance between the two teaching methods at the 0.01 significance level.

Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.


## 90% Confidence Interval for the True Population Mean

Given data:
- Population mean (\( \mu \)): 60 (used for comparison, not for CI calculation)
- Population standard deviation (\( \sigma \)): 8
- Sample size (\( n \)): 50
- Sample mean (\( \bar{x} \)): 65
- Confidence level: 90% (z-score ≈ 1.645)

### Confidence Interval Calculation

1. **Calculate the standard error (SE):**
   \[
   SE = \frac{\sigma}{\sqrt{n}}
   \]
   \[
   SE = \frac{8}{\sqrt{50}} \approx 1.13
   \]

2. **Calculate the margin of error (ME):**
   \[
   ME = z \times SE
   \]
   \[
   ME = 1.645 \times 1.13 \approx 1.86
   \]

3. **Compute the confidence interval:**
   \[
   CI = \bar{x} \pm ME
   \]
   \[
   CI = 65 \pm 1.86
   \]
   \[
   CI = (63.14, 66.86)
   \]

### Conclusion

The 90% confidence interval for the true population mean is approximately (63.14, 66.86).

Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

## Hypothesis Test for Caffeine Effect on Reaction Time

We will conduct a t-test to determine if caffeine has a significant effect on reaction time at a 90% confidence level.

### Given Data

- Sample size (\( n \)): 30
- Sample mean (\( \bar{x} \)): 0.25 seconds
- Sample standard deviation (\( s \)): 0.05 seconds
- Population mean (\( \mu \)): Assumed to be 0.3 seconds (standard reaction time without caffeine)

### Hypotheses

- Null hypothesis (\( H_0 \)): Caffeine does not have a significant effect on reaction time (\( \mu = 0.3 \)).
- Alternative hypothesis (\( H_1 \)): Caffeine has a significant effect on reaction time (\( \mu \neq 0.3 \)).

### T-Test Calculation

1. **Calculate the t-score:**
   \[
   t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
   \]
   \[
   t = \frac{0.25 - 0.3}{\frac{0.05}{\sqrt{30}}} = \frac{-0.05}{0.0091} \approx -5.48
   \]

2. **Degrees of freedom (df):**
   \[
   df = n - 1 = 30 - 1 = 29
   \]

3. **Determine the critical t-value for a two-tailed test at 90% confidence level:**
   \[
   t_{critical} \approx \pm 1.699 (from t-distribution table for df = 29)
   \]

4. **Compare t-score with critical t-value:**
   \[
   t = -5.48 < -1.699
   \]

### Conclusion

Since the calculated t-score (-5.48) is less than the critical t-value (-1.699), we reject the null hypothesis. This indicates that caffeine has a significant effect on reaction time at the 90% confidence level.
