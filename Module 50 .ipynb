Q4- What is the difference between AI, ML, DL, and DS?


Artificial Intelligence (AI): This is the idea of making machines do things that would require intelligence if done by humans.

Machine Learning (ML): A branch of AI where computers learn from data to make predictions or decisions without being directly programmed to perform the task.

Deep Learning (DL): A type of machine learning that uses complex networks of algorithms to analyze data, often performing well with large amounts of data.

Data Science (DS): This field uses scientific techniques to analyze data and extract insights. It includes AI and machine learning along with traditional statistics to address complex questions.

Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?


Supervised Learning: In this type of machine learning, the model is trained on a labeled dataset. This means that each input data point is paired with an output label. The model learns to predict the output from the input data. Common applications include classification and regression tasks.

Unsupervised Learning: This approach is used when the dataset does not include any labels. The model tries to learn the underlying patterns and structure from the data without any explicit instructions on what outputs are correct. Common methods include clustering and association to discover structure, detect anomalies, or reduce dimensions.

Semi-Supervised Learning: Semi-supervised learning falls between supervised and unsupervised learning. It uses a small amount of labeled data along with a larger amount of unlabeled data. This method helps in improving learning accuracy with less manually labeled data. It is useful when labeling data is expensive or time-consuming.

Q6: What are the training, testing, and validation splits? Explain the importance of each term.


Training Split: This portion of the dataset is used to train the machine learning model. It includes both input data and the corresponding output labels. The model learns to predict outputs from inputs based on this data. The size of the training set is typically the largest, as it directly influences the model's ability to learn.

Validation Split: This subset of data is used to tune the parameters of a model and prevent overfitting. It helps in validating the performance of the model during the training phase, allowing adjustments to model configurations (like hyperparameters). The validation set acts as a proxy for the test set during training.

Testing Split: After the model has been trained and validated, the testing split is used to assess its performance. This data is never seen by the model during training and acts as new, unseen data. The performance on the test set gives an unbiased evaluation of the final model fit on the training dataset.



Q7: How can unsupervised learning be used in anomaly detection?


Learning Normal Patterns: Unsupervised models are trained on data presumed to be mostly normal. These models learn the typical patterns and behaviors expected in the dataset.

Identifying Deviations: Once the model has learned what normal data looks like, it can identify deviations or outliers from these patterns. Anything that significantly differs from the learned norms can be flagged as an anomaly.

Cluster Analysis: Techniques like clustering can group similar data points together. Points that do not fit well into any cluster may be considered anomalies.

Dimensionality Reduction: Methods like PCA (Principal Component Analysis) can be used to reduce the complexity of data and highlight features that contribute most to its variance. Anomalies often appear as data points that have unusual values in these principal components.



Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.


Supervised Learning Algorithms:
- Linear Regression: Used for predicting numeric values.
- Logistic Regression: Used for binary classification tasks.
- Support Vector Machines (SVM): Effective in high-dimensional spaces.
- Decision Trees: Useful for both classification and regression.
- Random Forests: An ensemble method using multiple decision trees to improve prediction accuracy.
- Gradient Boosting Machines (GBM): Another powerful ensemble method for high accuracy in predictive tasks.
- Neural Networks: Highly versatile and capable of modeling complex non-linear relationships.

Unsupervised Learning Algorithms:

- K-Means Clustering: A method to partition data into K distinct groups based on feature similarities.
- Hierarchical Clustering: Builds a tree of clusters and is useful for hierarchical grouping.
- Principal Component Analysis (PCA): Reduces dimensionality while retaining most variability in the data.
- Autoencoders: Neural networks used for learning efficient codings of unlabeled data.
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Identifies clusters of varying shapes in data with noise.
- Gaussian Mixture Models (GMM): A probabilistic model assuming all data points are generated from a mixture of a finite number of Gaussian distributions with unknown parameters.
