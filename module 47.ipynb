Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
import scipy.stats as stats

# Observed frequencies
observed = [22, 28, 25, 15, 10, 20]

# Total number of M&Ms observed
total_observed = sum(observed)

# Expected frequencies based on the given distribution
expected = [total_observed * 0.20, total_observed * 0.20, total_observed * 0.20, total_observed * 0.10, total_observed * 0.10, total_observed * 0.20]

# Perform the chi-square goodness of fit test
chi2, p = stats.chisquare(observed, expected)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

if p < 0.05:
    print("Reject the null hypothesis: The observed distribution does not match the expected distribution.")
else:
    print("Do not reject the null hypothesis: The observed distribution matches the expected distribution.")
```

```python
import scipy.stats as stats

# Contingency table
data = [
    [20, 15],
    [10, 25],
    [15, 20]
]

# Perform the chi-square test
chi2, p, dof, expected = stats.chi2_contingency(data)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

if p < 0.05:
    print("Reject the null hypothesis: There is a significant association between the groups and outcomes.")
else:
    print("Do not reject the null hypothesis: There is no significant association between the groups and outcomes.")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
import math

# Given values
n = 500
x = 60
p_hat = x / n
z_value = 1.96  # Z-value for 95% confidence

# Calculate the standard error
standard_error = math.sqrt(p_hat * (1 - p_hat) / n)

# Calculate the margin of error
margin_of_error = z_value * standard_error

# Calculate the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```

Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.

```python
import scipy.stats as stats
import math

# Given values
mean = 75
std_dev = 12
sample_size = 30

# Z-value for 90% confidence
z_value = stats.norm.ppf(0.95)  # 90% confidence level corresponds to 95th percentile for one-tailed

# Calculate the margin of error
margin_of_error = z_value * (std_dev / math.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
dof = 10

# Generate a range of x values
x = np.linspace(0, 30, 500)

# Generate the chi-square distribution for the given degrees of freedom
y = stats.chi2.pdf(x, dof)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square distribution with {dof} degrees of freedom')

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 500)
y_shade = stats.chi2.pdf(x_shade, dof)
plt.fill_between(x_shade, y_shade, alpha=0.3, color='red', label='Area for chi-square statistic = 15')

# Add labels and title
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

```python
import math
from scipy import stats

# Given data
n = 1000  # sample size
p_hat = 520 / 1000  # sample proportion
confidence_level = 0.99  # confidence level

# Z-score for 99% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Standard error calculation
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error calculation
margin_of_error = z_score * standard_error

# Confidence interval calculation
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

```python
### Chi-Square Goodness of Fit Test

#### Hypotheses:
#- Null hypothesis (\(H_0\)): The coin is fair, so the expected frequencies of heads and tails are equal.
#- Alternative hypothesis (\(H_1\)): The coin is biased, meaning the observed frequencies will differ from the expected frequencies.

#### Data:
#- Observed frequencies: 45 tails and 55 heads.
#- Expected frequencies for a fair coin: 50 tails and 50 heads.

#### Chi-square test calculation:


import scipy.stats as stats

# Observed frequencies
observed = [45, 55]  # 45 tails and 55 heads

# Expected frequencies for a fair coin
expected = [50, 50]

# Calculate chi-square statistic and p-value
chi2_stat, p_val = stats.chisquare(f_obs=observed, f_exp=expected)

# Degrees of freedom
df = len(observed) - 1

# Critical value for 0.05 significance level
critical_value = stats.chi2.ppf(0.95, df)

chi2_stat, p_val, critical_value
```

Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.

### Chi-Square Test for Independence

#### Hypotheses:
- **Null hypothesis (\(H_0\))**: There is no association between smoking status and lung cancer diagnosis.
- **Alternative hypothesis (\(H_1\))**: There is an association between smoking status and lung cancer diagnosis.

#### Data:
The contingency table is:

|              | Lung Cancer: Yes | Lung Cancer: No |
|--------------|-------------------|-----------------|
| Smoker       | 60                | 140             |
| Non-smoker   | 30                | 170             |

#### Steps to Perform the Chi-Square Test for Independence:

1. **Calculate the expected frequencies** for each cell in the contingency table using the formula:
   \[
   E_{ij} = \frac{(row\ total \times column\ total)}{grand\ total}
   \]
   where \( E_{ij} \) is the expected frequency for cell \( i,j \).

2. **Calculate the chi-square statistic** using the formula:
   \[
   \chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
   \]
   where \( O_{ij} \) is the observed frequency for cell \( i,j \).

3. **Determine the degrees of freedom**:
   \[
   df = (number\ of\ rows - 1) \times (number\ of\ columns - 1)
   \]

4. **Find the critical value** for the chi-square statistic from the chi-square distribution table at the 0.05 significance level.

5. **Compare the chi-square statistic** to the critical value, or use the p-value to make the decision.

#### Calculation:

1. **Expected Frequencies**:

   \[
   \begin{aligned}
   &E_{11} = \frac{(90 \times 200)}{400} = 45 \\
   &E_{12} = \frac{(90 \times 200)}{400} = 45 \\
   &E_{21} = \frac{(310 \times 200)}{400} = 155 \\
   &E_{22} = \frac{(310 \times 200)}{400} = 155 \\
   \end{aligned}
   \]

2. **Chi-Square Statistic**:

   \[
   \chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
   \]

   \[
   \chi^2 = \frac{(60 - 45)^2}{45} + \frac{(140 - 155)^2}{155} + \frac{(30 - 45)^2}{45} + \frac{(170 - 155)^2}{155}
   \]

   \[
   \chi^2 = \frac{15^2}{45} + \frac{(-15)^2}{155} + \frac{(-15)^2}{45} + \frac{15^2}{155}
   \]

   \[
   \chi^2 = \frac{225}{45} + \frac{225}{155} + \frac{225}{45} + \frac{225}{155}
   \]

   \[
   \chi^2 = 5 + 1.45 + 5 + 1.45
   \]

   \[
   \chi^2 = 12.9
   \]

3. **Degrees of Freedom**:

   \[
   df = (2 - 1) \times (2 - 1) = 1
   \]

4. **Critical Value** at 0.05 significance level for 1 degree of freedom:

   \[
   \chi^2_{critical} = 3.841
   \]

#### Interpretation:
- **Chi-square statistic**: 12.9
- **Critical value**: 3.841

Since the chi-square statistic (12.9) is greater than the critical value (3.841), we reject the null hypothesis.

### Conclusion:
There is a significant association between smoking status and lung cancer diagnosis.

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

### Hypothesis Test for Population Mean

#### Hypotheses:
- **Null hypothesis (\(H_0\))**: The population mean is 70 (\(\mu = 70\)).
- **Alternative hypothesis (\(H_1\))**: The population mean is not 70 (\(\mu \neq 70\)).

#### Given Data:
- Sample size (\(n\)): 30
- Sample mean (\(\bar{x}\)): 72
- Sample standard deviation (\(s\)): 10
- Population mean under null hypothesis (\(\mu_0\)): 70
- Significance level (\(\alpha\)): 0.05

#### Steps to Perform the Hypothesis Test:

1. **Calculate the test statistic (t-value)** using the formula:
   \[
   t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}
   \]

2. **Determine the degrees of freedom (df)**:
   \[
   df = n - 1
   \]

3. **Find the critical value** for the t-distribution at the 0.05 significance level for a two-tailed test.

4. **Compare the test statistic** to the critical value, or use the p-value to make the decision.

#### Calculation:

1. **Test Statistic (t-value)**:

   \[
   t = \frac{72 - 70}{\frac{10}{\sqrt{30}}}
   \]

   \[
   t = \frac{2}{\frac{10}{5.477}}
   \]

   \[
   t = \frac{2}{1.826}
   \]

   \[
   t \approx 1.095
   \]

2. **Degrees of Freedom (df)**:

   \[
   df = 30 - 1 = 29
   \]

3. **Critical Value** for a two-tailed test at the 0.05 significance level with 29 degrees of freedom:

   From the t-distribution table, the critical value for \( \alpha/2 = 0.025 \) for 29 degrees of freedom is approximately \( \pm 2.045 \).

4. **Decision Rule**:
   - If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis.
   - Otherwise, we fail to reject the null hypothesis.

#### Interpretation:
- **Test statistic (t-value)**: 1.095
- **Critical value**: \( \pm 2.045 \)

Since the absolute value of the test statistic (1.095) is less than the critical value (2.045), we fail to reject the null hypothesis.

### Conclusion:
There is not enough evidence to suggest that the population mean is significantly different from 70 at the 0.05 significance level.
