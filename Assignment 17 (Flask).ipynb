Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a python framework that provides capabilities of creating web applications and allow other applications to connect and process data over it.  It provides a simple and flexible way to handle web requests, route URLs to functions, generate dynamic HTML pages using templates, integrate with databases, and more.

Advantages:

- Simplicity and ease of use.
- Lightweight and flexible.
- Extensible with a wide range of extensions.
- Efficient routing and URL mapping.
- Templating with Jinja2 for dynamic HTML generation.
- Built-in support for testing.
- Active community and comprehensive documentation.

Q3. What is App routing in Flask? Why do we use app routes?


In Flask, app routing refers to the process of mapping URLs to specific functions or view handlers in your application. It allows you to define the behavior of your application based on the URL that the user requests.

We use app routes in Flask to define the different endpoints or pages of our web application. Each route corresponds to a specific URL that a user can visit. By using app routes, we can associate these URLs with functions that will handle the requests and generate the appropriate response.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return("Welcome to ABC Corporation")

@app.route('/')
def info():
    return("Company Name: ABC Corporation\n Location: India Contact Detail: 999-999-9999")

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. It is a built-in function that helps generate URLs for routes defined in your application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'Contact us at example@example.com.'

if __name__ == '__main__':
    with app.test_request_context():

        home_url = url_for('home')
        about_url = url_for('about')
        contact_url = url_for('contact')

        print(home_url)     
        print(about_url)    
        print(contact_url)  
```
